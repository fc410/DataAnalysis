# Sales Analysis

#### Import Necessary Libraries

```python
import pandas as pd
import os
```

#### Merging 12 months of sales data into a single csv file

```python
df = pd.read_csv('Sales_data/Sales_April_2019.csv')
df.head(5)
```

```python
#reading all files in a directory
files = [file for file in os.listdir('Sales_Data')]
```

```python
#outputing the files in files
for file in files:
    print(file)
```

```python
#concatenate all the file into one single csv file
all_data = pd.DataFrame()

for file in files:
    df = pd.read_csv('Sales_data/'+file)
    all_data = pd.concat([all_data, df])
    
all_data.to_csv('all_data.csv', index=False)
    
all_data.head()
```

#### augmenting the data with aditional columns

```python
#adding a month column
#taking the first two characters from order date
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data.head()
```

# Clean up the Data!

```python
all_data.head(10)
```

```python
#dropping the second row
all_data.drop([1])
```

```python
#checking to see if any rows contain nan
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

```python
#dropping all the rows with nan
all_data = all_data.dropna(how='all')
all_data.head()
```

#### Find 'or' and delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
all_data.head()
```

```python
#converting the data in the Month column to integer
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

#### convert columns to the correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #make int

all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) #make float
```

#### Add a sales column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

# Question 1: What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month').sum()
```

So we can see that the best month for sale was December and the worst month
was January

```python
#visualizing the graph
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results['Sales'])
```

```python
plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD in Millions ($)')
plt.xlabel('Month Number')
plt.show()
```

visual representation that December had the best sales

# Question 2: What City had the highest number of sales

##### More augmentation

```python
#add a city column
#get the city from the purchase address column
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x)+ ', ' + get_state(x))

all_data.head(10)
```

```python
results = all_data.groupby('City').sum()
results
```

```python
#cities = all_data['City'].unique()

cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation='vertical', size=8)
plt.ylabel('Sales in USD in Millions ($)')
plt.xlabel('City Names')
plt.show()
```

With the statistics and the visual representation of that data we can conclude 
that San Francisco, CA had the highest number of sales

# Question 4: What products are most often sold together?

```python
#getting all the orders with duplicate order id
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df.head()
```

```python
#creating a new column with both products of the same order Id in one
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df.head()
```

```python
#droping the duplicates for better reading
df = df[['Order ID',  'Grouped']].drop_duplicates()
df.head(15)
```

```python
#couting all the pairs that accur together
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    
print(count)
```

```python
#more readable format
for key,value in count.most_common(10):
    print(key, value)
```

So we can see that the two items sold the most together would be an iphone and a lightning charging cable with 1005 of those two items sold

```python
count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 3)))
    
for key,value in count.most_common(10):
    print(key, value)
```

Most common three in a row would be a Google Phone, USB-C charging cable, and Wired Headphones

# Question 5: What product sold the most? Why do you think it sold the most?
