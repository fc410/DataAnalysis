# Filtering Data

```python
#getting a pokemon that is both grass and poison
#df.loc[(df['Type 1'] == 'Grass') & (df['Type 2']== 'Poison')]

poison_grass = df.where((df['Type 1'] == 'Grass') & (df['Type 2']== 'Poison')).dropna()
poison_grass
# Save csv file to poisonous-grass
poison_grass.to_csv('datasets/Pokemon/poison_grass_types.csv')
```

```python
#getting a pokemon that is grass and poison type with an hp greater than 70
new_df = df.where((df['Type 1'] == 'Grass') & (df['Type 2']== 'Poison') & (df['HP'] > 70)).dropna()
new_df

# rest the index
new_df = new_df.reset_index(drop=True)

new_df
```

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2']== 'Poison')& (df['HP'] > 70)]
```

```python
#Getting all the names that contain mega in their name
df.loc[(df['Name'].str.contains('Mega'))]
```

```python
#filtering out all the names that contain mega
df.loc[(~df['Name'].str.contains('Mega'))]
```

```python
#import regular expressions
import re

#getting a pokemon that is either fire or grass
df.loc[df['Type 1'].str.contains('Fire|Grass', regex=True)]
```

```python
#getting a pokemon that is either fire or grass without casing mattering
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
```

```python
#getting all the names that begin with the letters pi from the column 'Name'
#^-> emphasizes the beginning of the row
df.loc[df['Name'].str.contains('^pi[a-z]', flags=re.I, regex=True)]
```

# Aggregate Statistics (Groupby)

```python
#reloading the data frame
df = pd.read_csv('modified.csv')

df
```

```python
#sorting the means of the type 1 column with the highest defence in ascending 
#order
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

```python
#sorting the means of column type 1 in ascending order in the attack column 
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
#sorting the means of column type 1 in ascending order in relations to their HP
df.groupby(['Type 1']).mean().sort_values('HP', ascending=False)
```

```python
#summing up all the data for each column in relation to type 1
df.groupby(['Type 1']).sum()
```

```python
#counting the total
df.groupby(['Type 1']).count()
```

```python
#easier to read format for the count of all type 1 pokemons
df['Count'] = 1

df.groupby(['Type 1']).count()['Count']
```

```python
#can do multiple parameters
#really usefull for a data set that is massive
df.groupby(['Type 1', 'Type 2']).count()['Count']
```
