```python
# Import necessary libraries 
import pandas as pd
```

```python
# read the dataset & summarize it
test_results = pd.read_csv("datasets/TestAnalysis/test_scores.csv")
print(test_results.head())
print()

test_results.describe()
```

```python
# Set the index to the participant code
test_results.set_index('Participant_Code')

test_results.head()
```

```python
# Lets check how many null values there are
test_results.info()
```

#### Reading the data with pandas 

```python
test_results.head()
```

```python
# We want to drop the second row because it has not data 
test_results = test_results.drop(labels=1, axis=0)
test_results.head()
```

```python

test_results = test_results.fillna('X')
test_results.head()
```

```python
# Create a box plot for the post test scores of the children
is_post = test_results['Test'] == 'Post'
is_post = test_results[is_post]
is_post.plot(kind='box', x = 'Test', y = 'PercentGrade', title='Post Test Grades')
is_post.set_xlabel=('Post')
is_post.set_ylabel=('Grade')
```

```python
# Now create a box plot for the pre test scores and compare with the post
is_pre = test_results['Test'] == 'Pre'
is_pre = test_results[is_pre]
is_pre.plot(kind='box', x = 'Test', y = 'PercentGrade')
```

```python
print(is_post['PercentGrade'].mean())
print(is_pre['PercentGrade'].mean())
print(is_post['PercentGrade'].mean() - is_pre['PercentGrade'].mean())
```

As we can see there was more outliers in with the pre test scores than the post test cores. We can then inffer that students preformed better if they took a practice exam. Students preforms an average of about 8.41% better if they took a practice test which is almost a whole letter grade.

```python
# Graph both box plots in the same graph
results = test_results.drop(labels=0, axis=0)
results.boxplot(column='PercentGrade', by ='Test')
```
